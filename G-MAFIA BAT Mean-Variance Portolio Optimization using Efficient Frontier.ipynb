## 2.6 Covariance between Stocks

```python
#Covariance communicates the directional relationship between the returns on two assets. Positive covariance 
#indicates that assets move together (covariance value greater than 0). Negative covariance means they move inversely.
#(covariance value less than zero). Uncorrelated assets have no quantifiable pattern, and their covariance value is
#0. The covariance can be calculated with the given formula:
```

$$cov_{x,y}=\frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{N-1}$$

```python
#Generally, well-balanced portfolios reduce risk and volatility by pairing assets with negative covariance. If all 
#stocks across a portfolio are positively covariant, poor performance in one will mean some measure of poor performance 
#in the others, causing higher losses to the portfolio. (The opposite is also true, but losses are weighted more 
#heavily in human psychology because of loss aversion.)

#Below is a heatmap of covariance values between companies. (Covariance values are unstandardized and correlation 
#provides better indication of relationship strength between two variables, but covariance is sometimes preferred for
#similarly-scaled variables, like those in our current portfolio.)

covariance = daily_returns_summary.cov()
plt.figure(figsize=(16, 6))
cov_heatmap = sns.heatmap(covariance, annot=True, cmap='PuOr', annot_kws={"size": 12})
cov_heatmap.set_title('Covariance Heatmap', fontdict={'fontsize':14}, pad=10);
fig.tight_layout()
plt.savefig("cov_heatmap.png")
plt.show()
```

```python
#From these results, we can see that the BAT companies are slightly less covariant with the G-MAFIA companies as a
#whole, which makes sense, since they operate in somewhat different markets. Another standout is that
#IBM appears to be the least covariant with any other stock
```

## 2.7 Correlation between Stocks

```python
#Correlation is a similar concept to covariance, but the correlation coefficient is standardized, so that the values
#always fall between -1 and +1, making it more intuitive. Typically, it is better suited for instances when variables
#are scaled differently. Positive correlation indicates that assets move together (maximum value of +1). Negative 
#correlation means they move inversely (maximum value of -1). A correlation of 0 means no relation (they have no
#discernible effect on each other). The correlation can be calculated with the given formula (simplified):
```

$$\rho_{X,Y}=\frac{cov_{X,Y}}{\sigma_{X}\sigma_{Y}}$$

# 3. Mean-Variance Optimized Portolios

## 3.1 Setup

```python
#Mean-variance optimization is a method for maximizing the expected return of a portfolio of assets for a given level
#of risk. The high-level approach is to weight assets within a portfolio in different proportions, all adding up to 1,
#so that we have many diversified portfolio options. In basic language, this means deciding what percentage of total 
#money to hold in each company’s stock. We repeat this allocation process many times, and when we are finished, we 
#will have a body of options from which we can select for optimal return for any given risk level.

#In order to create and plot our portfolio options, we'll need to create two functions: one for generating x number of 
#portfolio options, and one for selecting the optimal portfolio from among those options, given a certain criterion.
#The first will be called return_portfolios and the second will be called optimal_portfolio:
```

```python
def return_portfolios(expected_returns, cov_matrix):
  np.random.seed(1)
  
  #(Reference: https://www.rubikpm.com/portfolio-optimization-with-python, 
  #https://www.machinelearningplus.com/machine-learning/portfolio-optimization-python-example/,
  #https://medium.com/@constandinou.antonio/quant-post-5-building-portfolios-along-the-efficient-frontier-e276de0a7adc)
  #set up arrays to hold our generated variables and results
  port_returns = []
  port_volatility = []
  stock_weights = []
    
  #here, we choose to generate 100,00 portfolios, with weights distributed 9 different ways each time. this process
  #involves linear algebraic tools like transpositions and dot products, filling and formatting the dataframe we will 
  #return
  selected = (expected_returns.axes)[0] 
  num_assets = len(selected) 
  num_portfolios = 100000
  for single_portfolio in range(num_portfolios):
    wts = np.random.random(num_assets)
    wts /= np.sum(wts)
    returns = np.dot(wts, expected_returns)
    volatility = np.sqrt(np.dot(wts.T, np.dot(cov_matrix, wts)))
    port_returns.append(returns)
    port_volatility.append(volatility)
    stock_weights.append(wts)
    
    portfolio = {'Returns': port_returns,
                 'Volatility': port_volatility}
  
  #finally, we package the information in a dataframe and return it
  for counter, symbol in enumerate(selected):
    portfolio[symbol +' Weight'] = [Weight[counter] for Weight in stock_weights]

  df = pd.DataFrame(portfolio)
  column_order = ['Returns', 'Volatility'] + [stock +' Weight' for stock in selected]
  df = df[column_order]
  return df
```

```python
def optimal_portfolio(returns):
    
    #this section is the headiest of our entire process. it's based on quadratic programming, linear algebra, and 
    #convex optimization, using the modern portfolio theory framework. there are various ways to accomplish the same 
    #goal, and a lot can be found online. to learn more, search "mean-variance optimization python," "efficient 
    #frontier python," or "cvxopt." (Some good starting resources: 
    #https://plotly.com/python/v3/ipython-notebooks/markowitz-portfolio-optimization/
    #http://cvxopt.org/examples/book/portfolio.html
    #https://cvxopt.org/userguide/coneprog.html, 
    #https://readthedocs.org/projects/cvxopt/downloads/pdf/latest/)
    
    n = returns.shape[1]
    returns = np.transpose(returns.values)

    N = 10
    mus = [10**(5.0 * t/N - 1.0) for t in range(N)] #mus represent expected asset returns—the object to be "minimized"

    # Convert to cvxopt matrices
    S = opt.matrix(np.cov(returns)) #covariance matrix of asset returns
    pbar = opt.matrix(np.mean(returns, axis=1))

    # Create constraint matrices
    G = -opt.matrix(np.eye(n))   # negative n x n identity matrix
    h = opt.matrix(0.0, (n ,1))
    A = opt.matrix(1.0, (1, n))
    b = opt.matrix(1.0)

    #Calculate efficient frontier weights using the cvxopt library's solver for linear and quadratic cone programs
    #qp = "quadratic programs"
    portfolios = [solvers.qp(mu*S, -pbar, G, h, A, b)['x'] for mu in mus]
    
    ##Calculate risks and returns for frontier using cvxopt BLAS (basic linear algebra subprograms) interface
    returns = [blas.dot(pbar, x) for x in portfolios]
    risks = [np.sqrt(blas.dot(x, S*x)) for x in portfolios]
    
    ##Calculate 2nd-degree polynomial of frontier curve
    m1 = np.polyfit(returns, risks, 2)
    x1 = np.sqrt(m1[2] / m1[0])
    
    #FINALLY! Calculate and return optimal portfolio
    wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']
    return np.asarray(wt), returns, risks
```

## 2.2 Efficient Frontier

```python
#The efficient frontier refers to the portfolios out of all those generated that yield the maximum return for a given
#level of acceptable volatility (or conversely, the volatility we must accept for a given expectation of returns). 
#It's easiest to understand through a graph visualization, with returns on the Y-axis and volatility on the X-axis.

#First, using our pre-build optimal_portfolio function, we plug in our daily_returns_summary to receive a list of the 
#optimal weights, returns, and risks:

wts, returns, risks = optimal_portfolio(daily_returns_summary[1:])
```

```python
#Below is a printout of total projected returns and volatility according to the weights from which each portfolio was
#composed, shown by company per row.

#Following that is a plot of all 100,000 portfolios generated. The top left edge of the data cloud represents all optimal 
#portfolios of stocks that maximize returns for the associated level of risk. The portfolios in the interior are 
#sub-optimal for their given risk level. For each of those, there is another that offers higher returns for the same 
#risk. The performance of each selected asset is plotted against volatility to show that, singularly, most do not
#reach the efficient frontier, but in composition, many of the plotted portfolios do.
```

```python
#Plug daily_mean and covariance into newly-created return_portfolios functions
portfolios = return_portfolios(daily_mean, covariance)
print(portfolios.head(10).round(4))

#Plotting the portfolio cloud
portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[10, 10], title = "Efficient Frontier")
plt.xlabel('Volatility (Std. Deviation)', fontsize = 14)

plt.ylabel('Expected Returns', fontsize = 14)
plt.title('Efficient Frontier', fontsize = 14)
plt.savefig("efficient_frontier.png")

single_asset_std = np.sqrt(np.diagonal(covariance))
plt.scatter(single_asset_std, daily_mean, marker='X',color='red',s=200)
for i, txt in enumerate(daily_returns_summary.keys()):
    plt.annotate(txt, (single_asset_std[i], daily_mean[i]), size=14, xytext=(10,10), ha='left', textcoords='offset points')

#Plotting the efficient portfolio
plt.plot(risks, returns, 'y-o')
plt.legend(['Efficient Frontier', 'Random'], fontsize = 20, loc = 'lower left')
plt.show()
```

## 2.3 Minimum Volatility Portfolio and Optimal Portfolio

```python
#The minimum volatility portfolio is the same as the left-most portfolio on the scatterplot. Any portfolio to its 
#right has increased risk.

#The optimal portfolio is found by determing which portfolio has the highest Sharpe ratio. The Sharpe ratio 
#represents the additional amount of return that an investor receives per unit of increase in risk. It is given below:
```

$$S = \frac{R_{p}-R_{f}}{\sigma_{p}}$$

$R_{p}$ = return of portfolio

$R_{f}$ = risk-free rate

$\sigma_{p}$ = standard deviation of portfolio

```python
#To find the portfolio with minimum volatility, .iloc allows for index selection to retrieve the entry that satisfies
#the requirement, while .idxmin points it toward the index with the lowest value for the volatility column (plotted
#below in red)
minimum_volatility_portfolio = portfolios.iloc[portfolios['Volatility'].idxmin()]                              
print(minimum_volatility_portfolio)

#To find the optimal portfolio, we create a temporary parameter of returns divided by volatility, find the highest 
#number with .idxmax, and return that entry with .iloc (plotted below in green)
optimal_portfolio = portfolios.iloc[((portfolios['Returns'])/portfolios['Volatility']).idxmax()]
print(optimal_portfolio)

plt.subplots(figsize=[10,10])
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(minimum_volatility_portfolio[1], minimum_volatility_portfolio[0], color='r', marker='o', s=500)
plt.scatter(optimal_portfolio[1], optimal_portfolio[0], color='g', marker='*', s=500)
plt.title("Minimum Volatility and Optimum Portfolios")
plt.xlabel("Volatility")
plt.ylabel("Returns")

plt.savefig("min_and_opt_portfolios.png")
```

```python
#Finally, investment decisions can be made according to multiple desires: minimum risk, highest Sharpe ratio, and
#highest return given specific risk criteria

#Closing observations:

#The minimum volatility portfolio is most heavily weighted toward IBM, while the optimal portfolio is most heavily 
#weighted toward Google

#The return on the minimum volatility portfolio is quite low as it is. Most investors want to trade some level of risk
#for higher potential reward to enjoy the rewards of stocks as an investment instrument.

#The volatility difference between the minimum volatility and optimal portfolios appears appears to be proportionally
#less than the associated increase in potential returns

#This mean-variance optimization demonstrates that while G-MAFIA BAT are all individual powerhouses, a portfolio derived
#from only them doesn't provide a very extended efficient frontier. Options for investors seem to be less than a more
#balanced portfolio, where risk tolerance can be adjusted with more impact. Balancing could involve choosing less 
#tech-centric companies and organizations not in direct competition with each other. Given the pandemic, the chosen time
#period may not be representative of performance in other circumstances, though all included stocks seem fairly reliable.
```
